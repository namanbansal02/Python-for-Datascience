```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# read the google 
import yfinance as yf
data_google = yf.download('Googl')
```

```python
data_google.head()
```

```python
data_google.tail()
```

### Simple Moving Average

```python
data_google['Open:10 days rolling'] = data_google['Open'].rolling(window=10,min_periods=1).mean()
```

```python
data_google.head()
```

```python
color = ['red','aqua']
data_google[['Open', 'Open:10 days rolling']].plot(figsize=(12,5), color=color)
```

### Cumulative Moving Average

```python
data_google['Open: CMA'] = data_google['Open'].expanding().mean()
```

```python
data_google.head()
```

```python
colors = ['red','aqua']
data_google[['Open','Open: CMA']].plot(figsize=(12,5), color=colors)
```

### Exponential Moving Average

```python
data_google['Open: EMA 10 days'] = data_google['Open'].ewm(span=10).mean()
```

```python
data_google.head()
```

```python
colors = ['red','aqua']
data_google[['Open','Open: EMA 10 days']].plot(figsize=(12,5), color=colors)
```

### Exponential Weighted Moving Average

```python
data_google['Open: EMA 10 days'] = data_google['Open'].ewm(alpha=0.5).mean()
```

```python
data_google.head()
```

```python
colors = ['red','aqua']
data_google[['Open','Open: EMA 10 days']].plot(figsize=(12,5), color=colors)
```
